```python
butter_prices = {
    "Almond": 10,
    "Peanut": 8,
    "Cashew": 12
}

fig, ax= plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Butter prices", ylabel="Price");
```

```python
fig, ax= plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()))
ax.set(title="Butter prices", ylabel="Price");
```

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("../introduction_to_pandas/car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2020", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True).astype(int)
car_sales
```

```python
car_sales
```

```python
car_sales["Sales Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Sales Date", y="Total Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter");
```
